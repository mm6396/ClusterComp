```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

dataset = pd.read_excel('test1.xlsx')  
X = dataset.iloc[:, [6, 7]].values  
df1 = pd.DataFrame(X, columns=['HARDNESS', 'MODULUS'])
print(df1)
plt.figure(figsize=(10,10))
plt.style.use("seaborn")
plt.title="hardness_vs_modulus"

plt.xlabel('Hardness (Ga)')  
plt.ylabel('Modulus  (Ga)')  
plt.legend() 

plt.scatter(df1["HARDNESS"],df1["MODULUS"],marker=".", s= 50 , c="blue")
plt.show()



df1["HARDNESS"] = pd.to_numeric(df1["HARDNESS"], downcast="float")
df1["MODULUS"] = pd.to_numeric(df1["MODULUS"], downcast="float")
df2 = pd.DataFrame(df1.dropna(), columns=['HARDNESS', 'MODULUS'])
df3=df2.dropna()

print(df3.head())




```

```python
k_means = KMeans(n_clusters=3,random_state=0 , max_iter=300, algorithm='auto').fit(df3)
y=k_means.predict(df3)
print(y)
df3['cluster'] = y
df3 = df3.apply(pd.to_numeric, errors='coerce')
print(df3)
u_labels = np.unique(y)
print("df3\n")
print(df3['HARDNESS'])
print(df3['cluster'])
# define and map colors
colors = ['#DF2020', '#81DF20', '#2095DF']
df3['c'] = df3.cluster.map({0:colors[0], 1:colors[1], 2:colors[2]})

plt.scatter(df3['HARDNESS'], df3['MODULUS'], marker='.', s=50, linewidths=3 , c = df3.c)
################################

```

```python
from fcmeans import FCM 
fcm = FCM(n_clusters=3 , Fuzzifier= 2)
fcm.fit(df3.to_numpy())

y = fcm.predict(df3.to_numpy())
print(y)
df3['cluster'] = y
df3 = df3.apply(pd.to_numeric, errors='coerce')
print(df3)
u_labels = np.unique(y)
print("df3\n")
print(df3['HARDNESS'])
print(u_labels)
colors = ['#DF2020', '#81DF20', '#2095DF']
df3['c'] = df3.cluster.map({0:colors[0], 1:colors[1], 2:colors[2]})
plt.scatter(df3['HARDNESS'], df3['MODULUS'], marker='.', s=100, linewidths=3, c=df3.c)


```
